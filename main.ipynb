# ANN model creation

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model =Sequential([#first hidden layer
                   Dense(units=10,
                         input_shape=(32*32*3,),
                         activation='sigmoid'),
                   #second hidden layer
                    Dense(units=10,
                         activation='sigmoid')

])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train_flat,y_train_flat,epochs=10)
```

```python
x_pred=model.predict(x_test_flat)
```

```python
x_pred
```

```python
model.evaluate(x_test_flat,y_test)
```
